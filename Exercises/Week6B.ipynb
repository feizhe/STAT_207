# SVD for Image Denoising

Given the following MNIST image with noise, how to denoise with SVD?

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the image
img = Image.open('mnist_data/15.png').convert('L')  # 'L' converts to grayscale

# Display the image
plt.imshow(img, cmap='gray')
plt.title("MNIST Digit")
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Convert to NumPy array
img_array = np.array(img)
img_noise = np.clip(img_array + np.random.normal(0, 30, img_array.shape), 0, 255)

plt.imshow(img_noise, cmap='gray')
plt.title("Noisy Image")
plt.axis('off')
plt.show()
```
